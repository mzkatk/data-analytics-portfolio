# Day 01-  Uber Rides Data Exploration
# Dataset: Uber rides NYC April 2014
# Goal: Load, explore, extract datetime features, observe patterns

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/plotly/datasets/master/uber-rides-data1.csv'

df = pd.read_csv(url, nrows=50000)
print(df.shape)
```

```python
df.head(10)
```

```python
print("shape:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nData Types:\n",df.dtypes)
print("\nNull Counts:\n",df.isnull().sum())
print("\nFirst 5 Rows:")
df.head(5)
print("\nDuplicates:", df.duplicated().sum())
```

- Dataset has 50k rows and 7 columns
- No null values found in any column
- 1324 Duplicates found
- Data is from April 2014

```python
df['Date/Time']=pd.to_datetime(df['Date/Time'])

df['hour'] = df['Date/Time'].dt.hour
df['day'] = df['Date/Time'].dt.day_name()
df['month'] = df['Date/Time'].dt.month
df['year'] = df['Date/Time'].dt.year

print(df.head())
print("\n")
print(df.dtypes)
```

```python
#basic stat on location
print(df[['Lat','Lon']].describe())
```

```python
mean = df[['Lat', 'Lon']].mean()
std = df[['Lat', 'Lon']].std()

outliers = df[
    ((df[['Lat', 'Lon']] < (mean - 3 * std)) |
     (df[['Lat', 'Lon']] > (mean + 3 * std)))
    .any(axis=1)
]

print(f"Outliers found: {len(outliers)}")
```

- Lat values range from 40.225900 to 373000
- Lon values range from -74.703900 to - 72.923700
- 2373 Outliers in location data

```python
# Busiest Hours
print(df.groupby('hour').size().sort_values(ascending=False).head(5))
```

```python
#Days with most ride
print(df.groupby('day').size().sort_values(ascending=False).head(3))
```

- Most rides happen at 17th hour
- Least rides at 2nd hour
- Wednesday has most rides: 11470
- Sunday has least rides: 3066
