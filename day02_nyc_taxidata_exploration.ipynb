```python
import pandas as pd

df = pd.read_csv(
    '/content/drive/MyDrive/datasets/yellow_tripdata_2015-01.csv',
    nrows=100000
)
print("shape:", df.shape)
print("\ncolumns:\n", df.columns.tolist())
df.head()
```

#Basic Data Exploration

```python
print("\nNull Count:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
print("\nData types:\n", df.dtypes)
df.describe()
#Data is already cleaned
```

#Fixing datetime column

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
#Trip Duaration
df['trip_duration_mins'] = round((df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60,2)
print(df[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'trip_duration_mins']].head())
```

#Extraction of time features

```python
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_day'] = df['tpep_pickup_datetime'].dt.day_name()
df['pickup_month'] = df['tpep_pickup_datetime'].dt.month
print(df[['tpep_pickup_datetime', 'pickup_hour', 'pickup_day', 'pickup_month']].head())

```

#Analysis

```python
#spotting impossible values
print("Fare amount min:", df['fare_amount'].min())
print("Trip distance min:", df['trip_distance'].min())
print("Passenger count min:", df['passenger_count'].min())
print("Trip duration min:", df['trip_duration_mins'].min())
print("\nFare amount max:", df['fare_amount'].max())
print("Trip distance max:", df['trip_distance'].max())
print("Passenger count max:", df['passenger_count'].max())
print("Trip duration max:", df['trip_duration_mins'].max())
```

```python
#bad rows
print("Negative fares:", len(df[df['fare_amount'] <= 0]))
print("Zero distance:", len(df[df['trip_distance'] <= 0]))
print("Zero passengers:", len(df[df['passenger_count'] <= 0]))
print("Negative duration:", len(df[df['trip_duration_mins'] <= 0]))
print("Too many passengers:", len(df[df['passenger_count'] > 6]))
```

```python
#Handling bad values
original = len(df)
df_clean = df[
    df['fare_amount'] > 0
    & (df['trip_distance'] > 0)
    & (df['passenger_count'] > 0)
    & (df['trip_duration_mins'] > 0)
]

print(f"Original rows: {original}")
print(f"Cleaned rows: {len(df_clean)}")
print(f"Removed rows: {original - len(df_clean)}")
```

#Outlier Detection

```python
mean_fare = df_clean['fare_amount'].mean()
std_fare = df_clean['fare_amount'].std()

outliers = df_clean[df_clean['fare_amount'] > mean_fare + 3*std_fare]

print(f"fare Outliers: {len(outliers)}")
print(f"Outlier Threshold: ${mean_fare + 3*std_fare:.2f}")
```

```python
#Quick Insights
print("Busiest hour:\n", df_clean.groupby('pickup_hour').size().sort_values(ascending=False).head(3))
print("\nBusiest day:\n", df_clean.groupby('pickup_day').size().sort_values(ascending=False))
print("\nAvg fare by passenger count:\n", df_clean.groupby('passenger_count')['fare_amount'].mean().round(2))
```
